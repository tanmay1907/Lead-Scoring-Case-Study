## Cleaning the dataset

##### Since there are values in the columns which are in proper case so converting all the values in lower case so that it may not effect our analysis.

```python
# Converting all the values to lower case
leads = leads.applymap(lambda s:s.lower() if type(s) == str else s)
```

##### There are values in the columns of the dataset which is 'select' which means the user has selected no option so we have to replace these values with NaN.

```python
# Replacing 'select' with NaN (Since it means no option is selected by the user.)
leads = leads.replace('select',np.nan)
```

```python
# Checking the uniqueness of the columns of the dataset.
leads.nunique()
```

##### Since there are columns in the dataset such as 'Magazine', 'Receive More Updates About Our Courses',etc. have one unique value only so it is better so drop these columns and dropping these columns won't affect our analysis.

```python
# Dropping the columns which have only one unique value.
leads= leads.drop(['Magazine','Receive More Updates About Our Courses','Update me on Supply Chain Content','Get updates on DM Content','I agree to pay the amount through cheque'],axis=1)
```

```python
# Now checking the percentage of missing values in the dataset.
(leads.isnull().mean()*100).sort_values(ascending=False)
```

##### Removing all the columns that are of no importance and also have missing values greater than 35% since 'Specialization' column is important  so we cannot remove it.

```python
# Removing columns which have missing values above 35% except Specialization column.
leads = leads.drop(['Asymmetrique Profile Index','Asymmetrique Activity Index','Asymmetrique Activity Score','Asymmetrique Profile Score','Lead Profile','Tags','Lead Quality','How did you hear about X Education','City','Lead Number'],axis=1)
```

```python
# Again checking the percentage of missing values.
(leads.isnull().mean()*100).sort_values(ascending=False)
```

```python
# Now checking the shape of the dataset.
leads.shape
```

##### There are very high number of missing values in 4 columns as seen above. But removing the rows with the null value will cost us a lot of data as they are important columns. So, instead we are going to replace the NaN values with 'not provided'. This way we have all the data and almost no null values. In case these come up in the model, it will be of no use and we can drop it off later.

```python
# Replacing NaN values of 'Specialization' column with 'not provided'
leads['Specialization'] = leads['Specialization'].fillna('not provided') 
```

```python
# Replacing NaN values of 'What matters most to you in choosing a course' column with 'not provided'
leads['What matters most to you in choosing a course'] = leads['What matters most to you in choosing a course'].fillna('not provided') 
```

```python
# Replacing NaN values of 'What is your current occupation' column with 'not provided'
leads['What is your current occupation'] = leads['What is your current occupation'].fillna('not provided') 
```

```python
# Replacing NaN values of 'Country' column with 'not provided'
leads['Country'] = leads['Country'].fillna('not provided') 
```

```python
# Now checking the info of the dataset.
leads.info()
```

```python
# Again checking the percentage of missing values.
(leads.isnull().mean()*100).sort_values(ascending=False)
```

```python
# # Checking the percent of data loss if the null values are removed
round(100*(sum(leads.isnull().sum(axis=1) > 1)/leads.shape[0]),2)
```

```python
# removing the null values which have values greater than 1.
leads = leads[leads.isnull().sum(axis=1) <1]
```

```python
# Again checking the percentage of missing values.
(leads.isnull().mean()*100).sort_values(ascending=False)
```

```python
# Checking value counts of all columns 
for column in leads:
    print(leads[column].astype('category').value_counts())
    print('----------------------------------------------------------------------------------------')
```

##### There is a column in the dataset which is 'Country' have have values of many countries and also not provided and majority is India so we can combine the values of other countries by the value 'outside india'

```python
# Checking value counts of country
leads["Country"].value_counts()
```

```python
# Replacing the values of countries except india and combining them by a value outside india.
def replace(x):
    category = ""
    if x == "india":
        category = "india"
    elif x == "not provided":
        category = "not provided"
    else:
        category = "outside india"
    return category

leads['Country'] = leads.apply(lambda x:replace(x['Country']), axis = 1)
```

```python
#  Again checking value counts of country
leads["Country"].value_counts()
```

```python
# Again checking value counts of all columns 
for column in leads:
    print(leads[column].astype('category').value_counts())
    print('----------------------------------------------------------------------------------------')
```

### Rescaling the Features 

##### As we can see in the dataset that all the variables have small integer values except Total time spent on website, Total Visits and Page Views per visit variables.So it is very important to rescale the variables so that all the variables have a comparable scale.If we don't do the rescaling then upon fiiting logistic regression model some of the coefficents obtained can be very large compared to other coefficients and this might create problem in model evaluation.So rescaling is very important and here we will do rescaling by MinMaxScaler.

```python
# For rescaling first of all we have to instantiate an object.
scaler=MinMaxScaler()
```

```python
# Applying the scaler to the numeriacl variables Total Time Spent on Website,Total Visits,Page Views Per Visit
num_vars=['Total Time Spent on Website','TotalVisits','Page Views Per Visit']
X_train[num_vars]=scaler.fit_transform(X_train[num_vars])
```

### Model Building

#### Recursive Feature Elimination(RFE)

```python
# Executing RFE with output variavle equal to 15.
logreg = LogisticRegression()
rfe = RFE(logreg, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
# checking the rank of the variables that have been selected by RFE
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# Checking the columns which are selected
col=X_train.columns[rfe.support_]
col
```

```python
# Checking the columns which are not selected
X_train.columns[~rfe.support_]
```

```python
# Selecting columns selected by RFE
X_train = X_train[col]
```

### Building model using Statsmodels

### Model 1

```python
# Adding a constant variable
X_train_sm = sm.add_constant(X_train)
```

```python
# Fitting the logistic regession model by using GLM
logm1 = sm.GLM(y_train, X_train_sm, family = sm.families.Binomial())
res=logm1.fit()
```

```python
# Let's see the summary of logistic regression model.
res.summary()
```

```python
# Importing 'variance_inflation_factor'
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Calculating VIF
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

##### We can see from the above model summary and VIF that VIF seems fine and but p value of Last Notable Activity_had a phone conversation feature is high so dropping it.

```python
X_train.drop('Last Notable Activity_had a phone conversation', axis = 1, inplace = True)
```

### Model 2

```python
# Refit the model with the new set of features
X_train_sm = sm.add_constant(X_train)
logm2 = sm.GLM(y_train, X_train_sm, family = sm.families.Binomial())
res = logm2.fit()
res.summary()
```

```python
# Calculating VIF
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

##### We can see from the above model summary and VIF that VIF seems fine and but p value of What is your current occupation_housewife feature is high so dropping it.

```python
# Dropping 'What is your current occupation_housewife feature
X_train.drop('What is your current occupation_housewife', axis = 1, inplace = True)
```

### Model 3

```python
# Refit the model with the new set of features
X_train_sm = sm.add_constant(X_train)
logm3 = sm.GLM(y_train, X_train_sm, family = sm.families.Binomial())
res = logm3.fit()
res.summary()
```

```python
# Calculating VIF
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

##### We can see from the modelm summary that all the VIF values are good and all the p-values are below 0.05. So we can fix model

### Making Predictions

```python
# Predicting the probabilities on the train set
y_train_pred = res.predict(X_train_sm)
y_train_pred[:10]
```

```python
# Reshaping to an array
y_train_pred = y_train_pred.values.reshape(-1)
y_train_pred[:10]
```

```python
#  Creating Data frame with given convertion rate and probablity of predicted ones
y_train_pred_final = pd.DataFrame({'Converted':y_train.values, 'Conversion_Prob':y_train_pred})
y_train_pred_final.head()
```

```python
# We select the random cut-off of 0.5 and Substituting 0 or 1 with it.
y_train_pred_final['Predicted'] = y_train_pred_final.Conversion_Prob.map(lambda x: 1 if x > 0.5 else 0)
y_train_pred_final.head()
```

### Model Evaluation

```python
# Importing metrics from sklearn for evaluation
from sklearn import metrics
```

```python
# Creating confusion matrix 
confusion = metrics.confusion_matrix(y_train_pred_final.Converted, y_train_pred_final.Predicted )
confusion
```

```python
# Predicted     not_churn    churn
# Actual
# not_churn        3438       457
# churn             747      1709
```

```python
# Check the overall accuracy
metrics.accuracy_score(y_train_pred_final.Converted, y_train_pred_final.Predicted)
```

##### We can see that that overall accuarcy is 81%.

```python
# Substituting the values of True Positives
TP = confusion[1,1]
```

```python
# Substituting the values of True Negatives
TN = confusion[0,0]
```

```python
# Substituting the value of False Positives
FP = confusion[0,1] 
```

```python
# Substituting the value of False Negatives
FN = confusion[1,0]
```

```python
# Now Calculating the sensitivity
TP/(TP+FN)
```

```python
# Calculating the specificity
TN/(TN+FP)
```

### Optimistic Cut-Off (ROC Curve)

##### We selected the cut-off of 0.5 randomly now we will find the optimum cut-off

```python
# First of all we define the ROC function
# ROC function
def roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
fpr, tpr, thresholds = metrics.roc_curve( y_train_pred_final.Converted, y_train_pred_final.Conversion_Prob, drop_intermediate = False )
```

```python
# Calling the roc function to draw ROC curve
roc(y_train_pred_final.Converted, y_train_pred_final.Conversion_Prob)
```

##### We can see from the above ROC curve that the area under ROC curve is 0.88 which is a very good value.
